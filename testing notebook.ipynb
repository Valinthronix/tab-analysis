```python
import tabula
import pandas as pd
```

```python
stevenson = tabula.read_pdf("C:\\Users\\jmeye\\Documents\\4 - Stevenson Regional Tab Summary 2018.pdf", java_options=["-Dfile.encoding=UTF8"], pages=[1,2], lattice=True)
# stevenson.shape
stevenson=stevenson.drop(stevenson.index[14]).reset_index(drop=True)
```

```python

def add_team_number(tab_df):
    try:
        team_numbers = []
        for team_string in tab_df["Team"]:
            team_number = team_string[:team_string.find("\r")]
            team_number = int(team_number)
            team_numbers.append(team_number)
        team_numbers = pd.Series(team_numbers, name="Team Number")
        tab_df.insert(0, "Team Number", team_numbers)
    except ValueError:
        pass

add_team_number(stevenson)
```

```python
def format_team_name(tab_df):
    for team_string_pair in tab_df["Team"].iteritems():
        team_string_pair[1]
        team_name = team_string_pair[1][team_string_pair[1].find("\r")+1:]
        tab_df.at[team_string_pair[0], "Team"] = team_name.replace("\r", " ")
        
format_team_name(stevenson)
```
