```python
import pandas as pd
import tabula
```

```python
st_paul_pg1 = tabula.read_pdf("stpaul.pdf", java_options=["-Dfile.encoding=UTF8"], pages=[1], lattice=True, pandas_options={'error_bad_lines' : False})
st_paul_pg2 = tabula.read_pdf("stpaul.pdf", java_options=["-Dfile.encoding=UTF8"], pages=[2], lattice=True, pandas_options={'error_bad_lines' : False})
```

```python
def fix_page(page_df):
    if page_df.shape[1] == 13:
        page_df.drop(columns=['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 4', 'Unnamed: 6', 'Unnamed: 8', 'Unnamed: 10', 'Unnamed: 12'], inplace=True)

fix_page(st_paul_pg1)
fix_page(st_paul_pg2)
```

```python


def countPages(pdfpath):
    import re
    rxcountpages = re.compile(rb"/Type\s*/Page([^s]|$)", re.MULTILINE|re.DOTALL)
    with open(pdfpath, 'rb') as file:
        data = file.read()
    return len(rxcountpages.findall(data))

countPages("stpaul.pdf")
```
